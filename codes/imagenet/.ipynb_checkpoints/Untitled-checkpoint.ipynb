```python
def fft(batch):
    batch = batch.unsqueeze(-1)
    B,C,H,W,Comlex = batch.shape
    batch_i = torch.zeros_like(batch)
    batch_complex = torch.cat((batch,batch_i),-1).float()
    batch_fr= torch.fft(batch_complex,2)
    return batch_fr.squeeze()
def permute(x):
    permute = [0, 1, 2]
    x = x / 255
    return x
def fft_filter(batch,percent=0.5,keep_low = True):
    if len(batch.shape) == 3:
        batch = batch.unsqueeze(0)
    batch = batch = batch.unsqueeze(-1)
    B,C,H,W,Comlex = batch.shape
    batch_i = torch.zeros_like(batch)
    batch_complex = torch.cat((batch,batch_i),-1).float()
    batch_fr= torch.fft(batch_complex,2)
    #B,C,H,W,Comlex
    H_cutoff = int(H * percent)
    for i in range(H_cutoff):
        W_percent = (percent**2 - i/H**2)**0.5
        W_percent = percent
        W_cutoff = int(W * W_percent)
        if keep_low:
            batch_fr[:,:,i:,W_cutoff:,:] = 0
        else:
            batch_fr[:,:,:i,:W_cutoff,:] = 0
    batch_filter = torch.ifft(batch_fr,2)[:,:,:,:,0].squeeze()
    return batch_filter
                
def fft_filter_quad(batch,percent=0.5,isquad=True):
    if percent >=1:
        return batch
    if len(batch.shape) == 3:
        batch = batch.unsqueeze(0)
    batch = batch = batch.unsqueeze(-1)
    B,C,H,W,Comlex = batch.shape
    batch_i = torch.zeros_like(batch)
    batch_complex = torch.cat((batch,batch_i),-1).float()
    batch_fr= torch.fft(batch_complex,2)
    #B,C,H,W,Comlex
    if isquad:
        #percent = 1 - percent
        percent /= 2
 
    H_cutoff = int(H * percent)
    W_cutoff = int(W * percent)
    if isquad:
        batch_fr[:,:,:,W_cutoff:W-W_cutoff,:] = 0
        batch_fr[:,:,H_cutoff:H-H_cutoff,:,:] = 0
        #batch_fr[:,:,0:H_cutoff,0:W_cutoff,:] = 0
        #batch_fr[:,:,0:H_cutoff,W-W_cutoff:W,:] = 0
        #batch_fr[:,:,H-H_cutoff:H,W-W_cutoff:W,:] = 0
        #batch_fr[:,:,H-H_cutoff:H,0:W_cutoff,:] = 0
 
    else:
        batch_fr[:,:,:,W_cutoff:,:] = 0
        batch_fr[:,:,H_cutoff:,:,:] = 0

    batch_filter = torch.ifft(batch_fr,2)[:,:,:,:,0].squeeze(-1)
    return batch_filter


def process_mean(x):
    x = x.squeeze()
    x = torch.abs(x)
    x = torch.mean(x,dim = 0)
    x = x.squeeze()
    x = torch.log(1+x)
    x = x/x.max()
    return x
def process_fft(x):
    x = x.squeeze()
    #x = x - x.min()
    x = torch.abs(x)
    x = torch.log(1+x)
    x = (x -x.min()) / (x.max() - x.min())
    #pdb.set_trace()
    return x
    
def imshow(img):
    plt.figure(figsize=(20,20))
    plt.imshow(img,cmap="gray")
    plt.show()
def cimshow(img):
    plt.figure(figsize=(20,20))
    plt.imshow(img)
    plt.show()
def dual_imshow(img1,img2):
    plt.figure(figsize=(20,10))
    plt.subplot(121)
    plt.imshow(img1)
    plt.subplot(122)
    plt.imshow(img2)
    plt.show()
    
def colshow(img):
    plt.figure(figsize=(20,20))
    plt.imshow(img)
    plt.show()
def ffr_pixel_delection(batch,deflections,window,rcam_prob=None,percent=0.9):
    if len(batch.shape) == 3:
        batch = batch.unsqueeze(0)
    batch = batch = batch.unsqueeze(-1)
    B,C,H,W,Comlex = batch.shape
    batch_i = torch.zeros_like(batch)
    batch_complex = torch.cat((batch,batch_i),-1).float()
    batch_fr= torch.fft(batch_complex,2)
    #B,C,H,W,Comlex
    B,C,H,W,Complex = batch_fr.size()
    #ramdom deflections
    de_std = deflections * 0.15
    deflections = int(np.random.normal(deflections,de_std))


    # percent = 1 - percent
    percent /= 2

    H_cutoff = int(H * percent)
    W_cutoff = int(W * percent)
    H_mid = int(H * 0.5)
    W_mid = int(W * 0.5)
    # batch_fr[:,:,:,W_cutoff:W-W_cutoff,:] = 0
    # batch_fr[:,:,H_cutoff:H-H_cutoff,:,:] = 0
    rcam_prob = np.zeros((H_mid,W))
    rcam_prob[:,W_cutoff:W-W_cutoff] = 1
    rcam_prob[H_cutoff:,:] = 1
    #rcam_prob = 1 - rcam_prob
    #pdb.set_trace()
    # rcam_prob = None
    def gauss_noise(img):
        std = torch.abs(img.clone().detach())*1
        #std = torch.ones_like(img) * 30
        # applied in low fre part
        #std[:,:,:,W_cutoff:W-W_cutoff,:] = 0
        #std[:,:,H_cutoff:H-H_cutoff,:,:] = 0
     
        # applied in high fre part

        std[:,:,0:H_cutoff,0:W_cutoff,:] = 0
        std[:,:,0:H_cutoff,W-W_cutoff:W,:] = 0
        std[:,:,H-H_cutoff:H,W-W_cutoff:W,:] = 0
        std[:,:,H-H_cutoff:H,0:W_cutoff,:] = 0
 

        img = img + torch.normal(0,std) 
        return img 
    # add gause noise
    # batch_fr = gauss_noise(batch_fr)
    #global map swap
    '''
    while deflections > 0:
        for c in range(C):
            x,y = randint(0,H-1),randint(0,W-1)

            if rcam_prob is not None:
                if uniform(0,1) > rcam_prob[x,y]:
                    continue

            while True:
                a,b = randint(-1 * window, window), randint(-1 * window, window)
                if x + a < H and x + a >0 and y+b < W and y+b > 0: break
            tmp = batch_fr[:,c,x,y,:].clone()
            batch_fr[:,c,x,y,:] = batch_fr[:,c , x + a, y + b,:]
            batch_fr[:,c , x + a, y + b,:] = tmp
            deflections -= 1
    '''

    '''
    #1/4 map swap
    while deflections > 0:
        for c in range(C):
        #if True:
            #c = [0,1,2]
            complex_channel = [0]
            x,y = randint(0,H_mid-1),randint(0,W_mid-1)

            if rcam_prob is not None:
                if uniform(0,1) > rcam_prob[x,y]:
                    continue

            while True:
                a,b = randint(-1 * window, window), randint(-1 * window, window)
                if x + a < H_mid and x + a >0 and y+b < W_mid and y+b > 0: break
            tmp = batch_fr[:,c,x,y,complex_channel].clone()
            batch_fr[:,c,x,y,complex_channel] = batch_fr[:,c , x + a, y + b,complex_channel]
            batch_fr[:,c , x + a, y + b,complex_channel] = tmp
            
            x = H - x
            a = -a 
            tmp = batch_fr[:,c,x,y,complex_channel].clone()
            batch_fr[:,c,x,y,complex_channel] = batch_fr[:,c , x + a, y + b,complex_channel]
            batch_fr[:,c , x + a, y + b,complex_channel] = tmp
            
            x = H - x
            a = -a
            y = W - y
            b = -b
            tmp = batch_fr[:,c,x,y,complex_channel].clone()
            batch_fr[:,c,x,y,complex_channel] = batch_fr[:,c , x + a, y + b,complex_channel]
            batch_fr[:,c , x + a, y + b,complex_channel] = tmp
            
            x = H - x
            a = -a
            tmp = batch_fr[:,c,x,y,complex_channel].clone()
            batch_fr[:,c,x,y,complex_channel] = batch_fr[:,c , x + a, y + b,complex_channel]
            batch_fr[:,c , x + a, y + b,complex_channel] = tmp
            deflections -= 4
    '''
    # '''
    #1/4 map replace
    clone_br = batch_fr.clone()
    
    while deflections > 0:
        #for c in range(C):
        if True:
            c = [0,1,2]
            complex_channel = [0]
            x,y = randint(1,H_mid-1),randint(1,W-1)

            if rcam_prob is not None:
                if uniform(0,1) > rcam_prob[x,y]:
                    continue

            while True:
                a,b = randint(-1 * window, window), randint(-1 * window, window)
                if x + a < H_mid and x + a >0 and y+b < W and y+b > 0: break
            batch_fr[:,c,x,y,complex_channel] = clone_br[:,c , x + a, y + b,complex_channel]
            # batch_fr[:,c , x + a, y + b,complex_channel] = clone_br[:,c,x,y,complex_channel]
            
        
            x = H - x
            a = -a
            y = W - y
            b = -b
            batch_fr[:,c,x,y,complex_channel] = clone_br[:,c , x + a, y + b,complex_channel]
            # batch_fr[:,c , x + a, y + b,complex_channel] = clone_br[:,c,x,y,complex_channel]
            

            deflections -= 2
    # '''
    
    #dual_imshow(itrans(clone_br[:,:,:,:,0].squeeze(),True),itrans(batch_fr[:,:,:,:,0].squeeze(),True))
    batch_filter = torch.ifft(batch_fr,2)[:,:,:,:,0].squeeze(-1)
    return batch_filter



```
