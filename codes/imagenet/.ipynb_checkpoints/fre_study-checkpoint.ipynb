```python
def imshow(img):
    plt.figure(figsize=(20,20))
    plt.imshow(img)
    plt.show()
def fft_filter(batch,percent=0.5,keep_low = True):
    if len(batch.shape) == 3:
        batch = batch.unsqueeze(0)
    batch = batch = batch.unsqueeze(-1)
    B,C,H,W,Comlex = batch.shape
    batch_i = torch.zeros_like(batch)
    batch_complex = torch.cat((batch,batch_i),-1).float()
    batch_fr= torch.fft(batch_complex,2)
    #B,C,H,W,Comlex
    H_cutoff = int(H * percent)
    for i in range(H_cutoff):
        W_percent = (percent**2 - i/H**2)**0.5
        W_percent = percent
        W_cutoff = int(W * W_percent)
        if keep_low:
            batch_fr[:,:,i:,W_cutoff:,:] = 0
        else:
            batch_fr[:,:,:i,:W_cutoff,:] = 0
                
    
    batch_filter = torch.ifft(batch_fr,2)[:,:,:,:,0].squeeze()
    return batch_filter
def fft_filter_quad(batch,percent=0.5,keep_low = True):
    if len(batch.shape) == 3:
        batch = batch.unsqueeze(0)
    batch = batch = batch.unsqueeze(-1)
    B,C,H,W,Comlex = batch.shape
    batch_i = torch.zeros_like(batch)
    batch_complex = torch.cat((batch,batch_i),-1).float()
    batch_fr= torch.fft(batch_complex,2)
    #B,C,H,W,Comlex
    H_cutoff = int(H * percent)
    W_cutoff = int(W * percent)
    batch_fr[:,:,:,W_cutoff:,:] = 0
    batch_fr[:,:,H_cutoff:,:,:] = 0

    batch_filter = torch.ifft(batch_fr,2)[:,:,:,:,0].squeeze()
    return batch_filter
def fft_round(batch,percent = 0.5):
    if len(batch.shape) == 3:
        batch = batch.unsqueeze(0)
    B,C,H,W = batch.shape
    #batch_i = torch.zeros_like(batch)
    #batch_complex = torch.cat((batch,batch_i),-1).float()
    batch_fr= torch.rfft(batch,2)
    batch_fl = torch.zeros_like(batch_fr)
    #pdb.set_trace()
    #B,C,H,W,Comlex
    H_cutoff = int(H * percent / 2)
    H_cutoff = H 
    H_mid_point = int(H / 2)
    W_mid_point = int(W / 2)
    '''
    #pdb.set_trace()
    for i in range(H_cutoff):
        #W_percent = (percent**2 - i/H_mid_point**2)**0.5
        W_percent = percent
        W_cutoff = int(W_mid_point * W_percent)
        #batch_fl[:,:,H_mid_point + i,W_mid_point - W_cutoff:W_mid_point + W_cutoff,:] = batch_fr[:,:,H_mid_point + i,W_mid_point - W_cutoff:W_mid_point + W_cutoff,:]
        #batch_fl[:,:,H_mid_point - i,W_mid_point - W_cutoff:W_mid_point + W_cutoff,:] = batch_fr[:,:,H_mid_point - i,W_mid_point - W_cutoff:W_mid_point + W_cutoff,:]
        batch_fl[:,:,i,W_mid_point - W_cutoff:W_mid_point + W_cutoff,:] = batch_fr[:,:,i,W_mid_point - W_cutoff:W_mid_point + W_cutoff,:]
        batch_fl[:,:,i,W_mid_point - W_cutoff:W_mid_point + W_cutoff,:] = batch_fr[:,:,i,W_mid_point - W_cutoff:W_mid_point + W_cutoff,:]


    '''
    #batch_fl = batch_fr
    #pdb.set_trace()
    H_cutoff = int(H * percent)
    #pdb.set_trace()
    for i in range(H_cutoff):
        W_percent = (percent**2 - i/H**2)**0.5
        W_percent = percent
        W_cutoff = int(W * W_percent)
        batch_fr[:,:,i:,W_cutoff:,:] = 0
    
    batch_filter = torch.irfft(batch_fr,2,signal_sizes = [H,W]).squeeze()
    #pdb.set_trace()
    return batch_filter
```
